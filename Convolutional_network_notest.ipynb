```python
import tensorflow as tf
```

## MaxPool2D

```python
x = tf.constant([[1., 2., 3.],
                 [4., 5., 6.],
                 [7., 8., 9.]])
```

```python
xx = tf.reshape(x, shape=(1,3,3,1)) #(batch, height, width, channels)
max_pool_2D = tf.keras.layers.MaxPool2D(pool_size=(3,3), padding='valid', strides=(1,1))
max_pool_2D(xx)
```

## ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread, imshow, subplots, show
```

```python
image = imread("image_test.jpg")
print(image.shape)
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
imshow(image)
```

```python
import matplotlib.pyplot as plt


image_generator_rotation = ImageDataGenerator(rotation_range=90)
image_generator_width = ImageDataGenerator(width_shift_range=0.4)
image_generator_hight = ImageDataGenerator(height_shift_range=0.4)
image_generator_britnes = ImageDataGenerator(brightness_range=(0.1, 0.9))

image_interator = image_generator_width.flow(images)
plt.subplot(2,2,1)
imshow(image_interator.next()[0].astype('int'))
image_interator = image_generator_rotation.flow(images)
plt.subplot(2,2,2)
imshow(image_interator.next()[0].astype('int'))
image_interator = image_generator_hight.flow(images)
plt.subplot(2,2,3)
imshow(image_interator.next()[0].astype('int'))
image_interator = image_generator_britnes.flow(images)
plt.subplot(2,2,4)
imshow(image_interator.next()[0].astype('int'))
```
